```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset (sales_category.csv should have columns: Category, Sales)
df = pd.read_csv("sales_category.csv")

# Plot bar chart with unique colors
plt.figure(figsize=(8,5))
plt.bar(df['Category'], df['Sales'], color=plt.cm.Paired(range(len(df))))

# Add labels and title
plt.xlabel("Product Category")
plt.ylabel("Sales")
plt.title("Sales Distribution by Product Category")

# Show chart
plt.show()

# Insights
best_category = df.loc[df['Sales'].idxmax(), 'Category']
print(f"Best-selling category: {best_category}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset (student_marks.csv should have columns: Subject, Average_Marks)
df = pd.read_csv("student_marks.csv")

# Plot bar chart
plt.figure(figsize=(8,5))
plt.bar(df['Subject'], df['Average_Marks'], color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.title("Average Marks of Students in Different Subjects")

# Show chart
plt.show()

# Insights
max_subject = df.loc[df['Average_Marks'].idxmax(), 'Subject']
min_subject = df.loc[df['Average_Marks'].idxmin(), 'Subject']
print(f"Easiest subject (highest average): {max_subject}")
print(f"Toughest subject (lowest average): {min_subject}")
```
